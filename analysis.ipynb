## Statistical analysis

```python
# Count number of wrong predictions

def get_error_counts(path: str):
    error_counts = []
    for i in range(1,51):
        sample = pl.scan_csv(f'{path}{i}.csv').with_columns(true_labels).collect()
        isCorrect_df = (sample['label'] == sample['true label']).value_counts()
        error_counts.append(isCorrect_df.filter(pl.col('label') == pl.lit(False)).select('count').item())
    return error_counts
```

```python
errors_dict = { 'datasets': ['baseline', 'L1', 'L2', 'L-inf'], 'error_counts': [] }

for path in csv_write_paths:
    errors_dict['error_counts'].append(get_error_counts(path))

errors = pl.from_dict(errors_dict)
errors.head()
```

```python
import pandas as pd
# loading the baseline DiFF Dataset
# filepath of saved csv version of dataset generated from HuggingFaceViT (New Dataset)
csv_save_dir = '/home/gscalubayan/diff-dataset/diff_df.csv'

def identifier_attrib(img_path):
    parts = Path(img_path).parts

    try:
        relative_path = parts[4:]
        if relative_path[0].lower() == "real":
            # e.g., Dataset/test/real/09048.png
            final_identifier = f'Dataset/{relative_path[2]}/real/{relative_path[-1]}'
        else:
            # e.g., Dataset/test/fake/FE/CoDiff/0C_interText_optDM_3_alpha=0.5.png
            final_identifier = f'Dataset/{relative_path[0]}/fake/{relative_path[1]}/{relative_path[2]}/{relative_path[-1]}'
        return final_identifier
    except Exception as e:
        print(f"Error with path: {img_path}\n{e}")
        return None
         
og_df = pd.read_csv(csv_save_dir)

og_df['identifier'] = og_df['image'].apply(identifier_attrib)

original_diff_ds = (
    HuggingFaceDataset.from_pandas(og_df)
    .cast_column('image', Image())
)
original_diff_ds[0]
```

```python
import os
def create_data_dict(data_path): # for fake images
    image_list = []
    category_list = []
    method_list = []
    split_list = []
    label_list = []
    identifier_list = []
    
    for root, _, files in os.walk(data_path):
        for file in files:
            full_path = os.path.join(root, file)
            parts = Path(full_path).parts  # safely get parts
            
            if len(parts) < 9:
                continue  # skip anything unexpected
                
            image_list.append(full_path)
            identifier_list.append("/".join(parts[7::]))
            if parts[9].lower() == "real":
                category_list.append("N/A")
                method_list.append("N/A")
                split_list.append(parts[8])
                label_list.append(parts[9])
            elif parts[9].lower() == "fake":
                category_list.append(parts[10])
                method_list.append(parts[11])
                split_list.append(parts[8])
                label_list.append(parts[9])
            else:
                continue # don't add anything
    lengths = [len(image_list), len(category_list), len(method_list), len(split_list), len(label_list), len(identifier_list)]
    print("Lengths:", lengths)
    return {
        'image': image_list,
        'category': category_list,
        'method': method_list,
        'split': split_list,
        'label': label_list,
        'identifier': identifier_list
    }

# Prepare datasets
adversarial_diff_l1_data_path = '/home/gscalubayan/cs199/huggingfacevit/DiFF_adversarial/fgsm_l1/Dataset'
adversarial_diff_l2_data_path = '/home/gscalubayan/cs199/huggingfacevit/DiFF_adversarial/fgsm_l2/Dataset'
#adversarial_diff_linf_data_path = sorted(Path('/home/gscalubayan/cs199/huggingfacevit/adversarial/fgsm_linf/Dataset').glob('*/*/*.*'))
```
